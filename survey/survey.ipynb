```python
import numpy as np
import pandas as pd
survey = pd.read_excel('Class Survey Responses.xlsx')
survey.head()
```

### countries bar chart

```python
countries = survey['Country of Origin']
# remove spaces
countries = countries.str.replace(' ', '')
countries = countries.str.lower()
count = countries.value_counts()
count.plot(kind='bar')
print(count)
```

# languages spoken

```python
languages = survey['Languages Spoken'].str.split(',')
languages = languages.apply(pd.Series).stack().reset_index(drop=True)
count = languages.value_counts()
count.plot(kind='bar')
print(count)
```
